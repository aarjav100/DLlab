```python
import tensorflow as tf
print(tf.__version__)
```

Create tensor

```python
#scaler tensor
a= tf.constant(5)
print(a)

#1d tensor
v=tf.constant([1,2,3,4])
print(v)

#2d tensor
m=tf.constant([[1,2],[3,4]])
print(m)

#3D tensor
t3= tf.constant([
 [[1, 2], [3, 4]],[[5, 6], [7, 8]]
])
print(t3)
```

Tensor Addition

```python
a= tf.constant([10,20,30])
b= tf.constant([1,2,3])
print(a+b)
```

checking datatype and shape of tensor

```python
x = tf.constant([1, 2, 3])
print(x.dtype, x.shape, tf.rank(x))
```

numpy array to tensor

```python
import numpy as np
np_array= np.array(([1,2,3],[4,5,6]))

tensor= tf.convert_to_tensor(np_array)
print(tensor)
```
